Bubble Sort

```python
def bubble(array):
    how_many = len(array)       # calculating the amount of stored data

    for out_loop in range (how_many):     # set the number of iterations according to the amount of data stored
        sorting = True      # mark the out_loop iteration as correct

        for inner_loop in range(how_many - out_loop - 1):       # mark each iteration of the out_loop
            if array[inner_loop] > array[inner_loop + 1]:       # compare the data in the current row with the data in the adjacent row
                array[inner_loop], array[inner_loop + 1] = array [inner_loop + 1], array[inner_loop]        # swap the larger variable on the right with the smaller variable on the left
                sorting = False         # returns false if a position swap occurs
        
        if sorting == True:     # create a condition if the sequence is correct
            break       
    return array

jack = 11
queen = 12
king = 13
ace = 1
ex = [5, 9, 10, ace, 6, 7, queen, 4, king, 2, jack, 3, 8]

print("original :", ex)
bubble(ex)
print("after sorting :", ex)
```

Insertion Sort

```python
def insertion(array):

    for repeat in range(1, len(array)):     # Count the number of data entries. The count starts at index 1 (the second entry).
        item_key = array[repeat]        # store a large amount of data 

        comparator = repeat - 1     # Determining the index of the element to the left of `item_key`

        while comparator >= 0 and array[comparator] > item_key:     # Check that the limit does not exceed index 0 and compare the element to the left
            array[comparator + 1] = array[comparator]       # Larger elements are copied or shifted one position to the right
            comparator -= 1     # Go to the previous number on the left to check it

        array[comparator + 1] = item_key        # reassign the previously saved value to the new one.
    return array

jack = 11
queen = 12
king = 13
ace = 1
example = [4 ,2 ,5 ,6 ,ace ,9 ,queen ,8 ,king ,10 ,jack ,7 ,3 ]
print("before :", example)
insertion(example)
print("after :", example)
```

Merge Sort

```python
def merge(left, right):
    if len(left) == 0: 
        return right    # If the left list is empty, return the right list
    if len(right) == 0:
        return left     # If the right list is empty, return the left list
    
    result = []     # A container for the merged results.
    index_l = index_r = 0       # Pointers indicating our position in the left and right lists.
    
    while len(result) < len(left) + len(right):     # The loop occurs if the number of data points in the `result` variable is fewer than the sum of the data points in the left and right parameters
        if left[index_l] <= right[index_r]:     # The code will be executed if the left side is less than or equal to the right side
            result.append(left[index_l])        # add the left side to the variable result
            index_l += 1        # shift the variable index
        else:
            result.append(right[index_r])      # add the right side to the variable result
            index_r += 1        # shift the variable index
            
        if index_r == len(right):
            result += left[index_l:]        # If the right list runs out first, the remaining elements in the left list are immediately appended to the result
            break
        if index_l == len(left):
            result += right[index_r:]       # If the left list runs out first, the remaining elements in the left list are immediately appended to the result
            break
    return result

def merge_sort(array):
    if len(array) < 2:      # Conditions when the list is empty or contains only one element
        return array
    
    mid = len(array) // 2       # finding the midpoint by dividing

    left_part = merge_sort(array[:mid])     # Split the left half and call this function again (recursively) until reach the base case.
    right_part = merge_sort(array[mid:])        # Split the right half and call this function again (recursively) until you reach the base case.
    
    return merge(left_part, right_part)

jack = 11
queen = 12
king = 13
ace = 1
example = [queen, 4, 8, 10, ace, 3, king, 6, 2, jack, 5, 9, 7]
print("before :", example)
example = merge_sort(example)
print("after  :", example)
```

Tugas

```python
def tugas(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr.pop() 
    kiri = []
    kanan = []
    for elemen in arr:
        if elemen <= pivot:
            kiri.append(elemen) 
        else:
            kanan.append(elemen)  
    return tugas(kiri) + [pivot] + tugas(kanan)

ex = [64, 34, 25, 12, 22, 11, 90]
print("before :", ex)
hasil = tugas(ex)
print("Hasil pengurutan:", hasil)
```
