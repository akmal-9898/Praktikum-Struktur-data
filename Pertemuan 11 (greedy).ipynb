# 1. Coin Change
- Detail : diberi sejumlah nilai yang dapat ditukar dengan koin yang mendekati nilai yang sams dengan nilai tersebut. Memilih koin dengan nilai terbesar yang masih bisa digunakan pada setiap langkah, lalu menguranginya dari jumlah yang masih harus dikembalikan

- Tujuan : mencari jumlah minimum koin untuk membentuk sebuah minimum tertentu

```python
def coin_change_greedy(money_to_coin, available_coins):
    available_coins.sort(reverse=True)    # mengurutkan koin dari yang terbesar hingga terkecil
    result = []   # menyimpan koin yang dipilih sebagai hasil
    for for_saving_coin in available_coins:   # Variabel for_saving_coin di sini adalah variabel loop yang akan bergantian berisi nilai 25, 10, 5, 1 (setelah di-sort).
        while money_to_coin >= for_saving_coin:   # Ini memungkinkan penggunaan koin yang sama lebih dari satu kali
            money_to_coin -= for_saving_coin    # memakai koin yang ditentukan lewat pengurangan
            result.append(for_saving_coin)    # Menambahkan koin yang digunakan ke dalam list result
    return result

money_to_coin = 100
available_coins = [25, 10, 5, 1]
change = coin_change_greedy(money_to_coin, available_coins)
print("uang yang ingin di konversi :", money_to_coin)
print("Koin yang tersedia :", available_coins)
print("Koin yang digunakan :", change)
```

# 2. Frantional Knapsack
- Detail : diberikan sejumlah barang. masing-masing dengan berat dan nilai serta sebuah ransel dengan kapasitas tertentu

- Tujuan : Mengisi ransel agar nilai total barang maksimal dengan kemungkinan mengambil sebagian dari barang jika diperlukan

```python
def fractional_knapsack (daftar_barang,max_capacity):   # setiap daftar barang berupa tuple (value dan berat)
  daftar_barang = sorted(daftar_barang, key=lambda x: x[0]/x[1], reverse=True)    # Mengurutkan barang berdasarkan nilai per berat (value/weight) dari yang tertinggi ke terendah.
  value_yang_berhasil_disimpan = 0.0
  for value_barang, berat_barang in daftar_barang:
      if max_capacity >= berat_barang:    # Jika kapasitas yang tersisa masih cukup untuk mengambil barang utuh, maka ambil seluruh barang.
          max_capacity -= berat_barang
          value_yang_berhasil_disimpan += value_barang
      else:
          value_yang_berhasil_disimpan += value_barang * (max_capacity / berat_barang)    # Ambil sebanyak mungkin yang masih muat (pecahan).
          break
  return value_yang_berhasil_disimpan

daftar_barang = [(60, 10), (100, 20), (120, 30)]
max_capacity = 50
max_value = fractional_knapsack(daftar_barang, max_capacity)
print("indikator ----- [(value,berat)]")
print("daftar barang :", daftar_barang)
print("kapasitas ransel :", max_capacity)
print("\nNilai maksimum yang dapat dibawa(value) :", max_value)
```

# 3. Minimum Spanning Tree : Prim
- Detail : Minimum Spanning Tree adalah salah satu aplikasi algoritma greedy dalam konteks graf. Diberikan seuah graf berbobot yang terhubung

- Tujuan : menghubungkan semua simpul dalam graf dengan total bobot minimum, tanpa membentuk siklus

- Prim : memulai dari 1 simpul dan secara bertahap menambahkan sisi (edge) dengan bobot terkecil yang menghubungkan simpul yang sudah dikunjungi dengan simpul yang belum

```python
import heapq
def prim_mst(graph, start):
    visited = set()
    min_heap = [(0, start)]
    total_weight = 0
    while min_heap:
        weight, node = heapq.heappop(min_heap)
        if node in visited:
            continue
        visited.add(node)
        total_weight += weight
        for neighbor, edge_weight in graph[node]:
            if neighbor not in visited:
                heapq.heappush(min_heap, (edge_weight, neighbor))
    return total_weight

graph = {
    'A': [('B', 2), ('C', 3)],
    'B': [('A', 2), ('C', 1), ('D', 1)],
    'C': [('A', 3), ('B', 1), ('D', 4)],
    'D': [('B', 1), ('C', 4)]
}

print("Total bobot MST (Prim):", prim_mst(graph, 'A'))
```

# 4. Minimum Spanning Tree : Kruskal
- Detail : Minimum Spanning Tree adalah salah satu aplikasi algoritma greedy dalam konteks graf. Diberikan seuah graf berbobot yang terhubung

- Tujuan : menghubungkan semua simpul dalam graf dengan total bobot minimum, tanpa membentuk siklus

- Kruskal : memulai dari semua sisi graf, lalu memilih sisi dengan bobot terkecil yang tidak membentuk siklus, hingga semua simpul terhubung. kedua pendekatan ini menunjukan bagaimana strategi greedy digunakan dalam membangun solusi secara bertahap dengan selalu memilih opsi lokal terbaik tiap langkah

```python
def kruskal_mst(edges, n_nodes):
    parent = {i: i for i in range(n_nodes)}

    def find(x):
        while parent[x] != x:
            x = parent[x]
        return x

    def union(x, y):
        root_x = find(x)
        root_y = find(y)
        if root_x != root_y:
            parent[root_y] = root_x
            return True
        return False
    edges.sort(key=lambda x: x[2])
    total_weight = 0
    for u, v, weight in edges:
        if union(u, v):
            total_weight += weight
    return total_weight

edges = [
    (0, 1, 2),  # A-B
    (0, 2, 3),  # A-C
    (1, 2, 1),  # B-C
    (1, 3, 1),  # B-D
    (2, 3, 4)   # C-D
]

print("Total bobot MST (Kruskal):", kruskal_mst(edges, 4))
```
