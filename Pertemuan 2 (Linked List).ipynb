1. Buatlah double linked list dengan 3 node didalamnya, kemudian lakukan penyisipan
tengah dan hapus dengan kata kunci !

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.previous = None
n1 = Node("NSX")
n2 = Node("Silvia")
n3 = Node("Cappucino")
class LinkedList:
    def __init__(self):
        self.the_head = None
    def listprint(self):
        head_pointer = self.the_head
        while head_pointer is not None:
            print(head_pointer.data)
            head_pointer = head_pointer.next
    def AddInBetween(self,the_middle,new_data):
        if the_middle is None:
            print("no data on the middle node")
            return
        new_node = Node(new_data)
        new_node.next = the_middle.next
        the_middle.next = new_node
    def RemoveNode(self, removekey):
        the_head = self.the_head
        if (the_head is not None):
            if (the_head.data == removekey):
                self.the_head = the_head.next
                the_head = None
                return
        while (the_head is not None):
            if the_head.data == removekey:
                break
            previous = the_head
            the_head = the_head.next
        if (the_head == None):
            return
        previous.next = the_head.next
        the_head = None
linkedlist = LinkedList()
linkedlist.the_head = n1
linkedlist.the_head.next = n2
n2.previous = n1
n2.next = n3
n3.previous = n2
print("=====linked list asli=====")
linkedlist.listprint()
print("=====setelah di tambah di tengah=====")
linkedlist.AddInBetween(n1,"Estilo")
linkedlist.listprint()
print("=====setelah di hapus dengan kata kunci=====")
linkedlist.RemoveNode("Silvia")
linkedlist.listprint()
```

2. Buatlah circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan
akhir dan hapus akhir !

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.previous = None
n1 = Node("NSX")
n2 = Node("Silvia")
n3 = Node("Cappucino")
class CircularDoubly:
    def __init__(self):
        self.the_head = None
    def listprint(self):
        if not self.the_head:
            return
        curr = self.the_head
        while True:
            print(curr.data)
            curr = curr.next
            if curr == self.the_head: 
                break
    def RemoveEnd(self):
        if self.the_head is None:
            return
        if self.the_head.next == self.the_head:
            self.the_head = None
            return
        tail = self.the_head.previous
        new_tail = tail.previous
        new_tail.next = self.the_head
        self.the_head.previous = new_tail
    def AddEnd(self, new_node):
        if not self.the_head:
            self.the_head = new_node
            new_node.next = new_node
            new_node.previous = new_node
            return
        tail = self.the_head.previous
        tail.next = new_node
        new_node.previous = tail
        new_node.next = self.the_head
        self.the_head.previous = new_node
linkedlist = CircularDoubly()
linkedlist.the_head = n1
n1.next = n2
n1.previous = n3 
n2.previous = n1
n2.next = n3
n3.previous = n2
n3.next = n1
print("=====linked list asli=====")
linkedlist.listprint()
print("=====setelah menambah node di ahir=====")
n4 = Node("WRX")
linkedlist.AddEnd(n4)
linkedlist.listprint()
print("=====setelah menghapus node di ahir=====")
linkedlist.RemoveEnd()
linkedlist.listprint()
```
