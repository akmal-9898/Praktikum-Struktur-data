stack using a list

```python
holy_trinity = []  # serves as a container
holy_trinity.append("918 spyder")  # to add data from stack
holy_trinity.append("P1")
holy_trinity.append("LaFerrari")
print(f"stack content :{holy_trinity}")

print(f"element removed from the stack : {holy_trinity.pop()}")  # to delete data from stack

print(f"after delete :{holy_trinity}")
```

stack using a Collections.deque

```python
from collections import deque

hypercar = deque()  # serves as a container
hypercar.append("chiron")   # to add data from stack
hypercar.append("jesko")
hypercar.append("venom")
print(f"stack content : {hypercar}")

print(f"element removed from the stack : {hypercar.pop()}") # to delete data from stack

print(f"after delete :{hypercar}")
```

stack using a queue.LifoQueue

```python
from queue import LifoQueue
JDM = LifoQueue(maxsize = 5)
JDM.put("NSX")
JDM.put("MK4")
JDM.put("R34")    # adding data by waiting
JDM.put_nowait("WRX")    # add data without waiting
JDM.put_nowait("RX7")
print(f"is the stack empty : {JDM.empty()}")    # checking if the stack is empty
print(f"Is the stack full : {JDM.full()}")    # checking whether the stack is full
print(f"how much data is on the stack : {JDM.qsize()}")    # checking the stack size

print(f"\ndeleting data by waiting : {JDM.get()}")
print(f"deleting data without waiting : {JDM.get_nowait()}")
print(f"deleting data without waiting : {JDM.get_nowait()}")
print(f"deleting data without waiting : {JDM.get_nowait()}")
print(f"deleting data without waiting : {JDM.get_nowait()}")

print(f"\nis the stack empty : {JDM.empty()}")    # checking if the stack is empty
print(f"Is the stack full : {JDM.full()}")    # checking whether the stack is full
print(f"how much data is on the stack : {JDM.qsize()}")    # checking the stack size
```

stack using a single linked list

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
class Stack:
    def __init__(self):
        self.head = Node("head")
        self.size = 0
    def __str__(self):
        cur = self.head.next
        out = ""
        while cur:
            out += str(cur.value) + " -> "
            cur = cur.next
        return out[:-2]
    def getSize(self):
        return self.size
    def isEmpty(self):
        return self.size == 0
    def peek(self):
        if self.isEmpty():
            raise Exception("value is empty")
        return self.head.next.value
    def push(self, value):
        node = Node(value)
        node.next = self.head.next
        self.head.next = node
        self.size += 1
    def pop(self):
        if self.isEmpty():
            raise Exception("The stack does not have data to perform the pop command.")
        remove = self.head.next
        self.head.next = self.head.next.next
        self.size -= 1
        return remove.value
if __name__ == "__main__":
    s = Stack()
    s.push("revuelto")
    s.push("12 cilindri")
    s.push("vanquish")
    print(f"stack contents : {s}") 
    print(f"top data : {s.peek()}") 
    print(f"deleted data : {s.pop()}")
    print(f"remaining data : {s}")
```
