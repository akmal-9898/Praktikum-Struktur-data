queue using list

```python
aston_martin = []

aston_martin.append("valhalla")
aston_martin.append("vanquish")
aston_martin.append("valkriye")
aston_martin.append("DBS")
print(f"data on queue(list) : {aston_martin}")

print(f"\nremove data from queue(list) : {aston_martin.pop(0)}")
print(f"remove data from queue(list) : {aston_martin.pop(0)}")
print(f"\nafter removing some data on queue(list) : {aston_martin}")
```

queue using collection deque

```python
from collections import deque

lamborghini = deque()

lamborghini.append("revuelto")
lamborghini.append("temerario")
lamborghini.append("hurracan")
lamborghini.append("aventador")
lamborghini.append("murcielago")
lamborghini.append("gallardo")
print(f"data on queue(deque) : {lamborghini}")

print(f"\nremove data from queue(deque) : {lamborghini.popleft()}")
print(f"remove data from queue(deque) : {lamborghini.popleft()}")
print(f"remove data from queue(deque) : {lamborghini.popleft()}")
print(f"\nafter removing some data on queue(list) : {lamborghini}")
```

queue using queue.LifoQueue

```python
from queue import Queue

porsche = Queue(maxsize = 6)
print(f"Is the data in the queue full(LifoQueue) : {porsche.full()}")   
print(f"data on queue(LifoQueue) : {porsche.qsize()}")
print(f"is the queue empty(LifoQueue) : {porsche.empty()}\n")
porsche.put("911")
porsche.put("718")
porsche.put("969")
porsche.put("918")
porsche.put("912")
porsche.put("922")
print(f"Is the data in the queue full(LifoQueue) : {porsche.full()}")
print(f"data on queue(LifoQueue) : {porsche.qsize()}")
print(f"is the queue empty(LifoQueue) : {porsche.empty()}")
print(f"\ntake data from queue(LifoQueue) : {porsche.get()}")
print(f"take data from queue(LifoQueue) : {porsche.get()}")
print(f"take data from queue(LifoQueue) : {porsche.get()}")
print(f"\nIs the data in the queue full(LifoQueue) : {porsche.full()}")
print(f"data on queue(LifoQueue) : {porsche.qsize()}")
print(f"is the queue empty(LifoQueue) : {porsche.empty()}")
print(f"\ntake data from queue(LifoQueue) : {porsche.get()}")
print(f"take data from queue(LifoQueue) : {porsche.get()}")
print(f"take data from queue(LifoQueue) : {porsche.get()}")
print(f"\nIs the data in the queue full(LifoQueue) : {porsche.full()}")
print(f"data on queue(LifoQueue) : {porsche.qsize()}")
print(f"is the queue empty(LifoQueue) : {porsche.empty()}")
```

queue using single linked list

```python
class Rally:
    def __init__(self):
        self.queue = list()
    
    def add_rally_car(self, data):
        if data not in self.queue:
            self.queue.insert(0, data)
            return True
        return False
    
    def remove_rally_car(self):
        if len(self.queue) > 0:
            return self.queue.pop()
        return ("No car in Queue!")
    
    def size_of_the_garage(self):
        return len(self.queue)
    
my_garage = Rally()
print(my_garage.add_rally_car("Mitsubishi Lancer"))
print(my_garage.add_rally_car("MINI john Cooper Works"))
print(my_garage.add_rally_car("Citroen C4"))
print(my_garage.add_rally_car("Citroen Xsara"))
print(my_garage.add_rally_car("Subaru Impreza"))
print(my_garage.add_rally_car("Volkswagen Polo R"))
print(f"\nqueue size (single-linked list) : {my_garage.size_of_the_garage()}")
print(f"deleted element : {my_garage.remove_rally_car()}")
```
