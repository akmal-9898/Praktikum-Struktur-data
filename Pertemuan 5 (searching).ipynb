- Linear Search

```python
def linear (array,element,target):

  for variabel in range(element):
      if (array[variabel] == target):
        return variabel
  return "element not found"

if __name__ == '__main__':
  array = [1,2,3,4,5]
  target = 1
  element = len(array)
print("index :",linear(array,element,target))
```

- Binary Search iteratif

```python
def binary (array,left,right,target):
  while 1 <= right:
    mid = left + (right - left) //2
    if array[mid] == target:
      return mid
    elif array[mid] < target:
      left = mid + 1
    else:
      right = mid - left
  return -1

if __name__ == "__main__":
  array = [10,20,30,40,50,60,70]
  target = 60

  if binary(array,0,len(array)-1,target) == -1:
    print("element not found")
  else :
    print(f"index : {binary(array,0,len(array)-1,target)}")
```

- Binary Search rekursif

```python
def binary(array, left, right, target):
    if right >= left:
        mid = left + (right - left) // 2
        
        if array[mid] == target:
            return mid
        elif array[mid] > target:
            return binary(array, left, mid - 1, target)
        else:
            return binary(array, mid + 1, right, target)
    else:
        return -1

if __name__ == "__main__":
    array = [10, 20, 30, 40, 50, 60]
    target = 40
    result = binary(array, 0, len(array) - 1, target)

    if result == -1:
        print("Element not found")
    else:
        print(f"Element found at index: {result}")
```

- Jump Search

```python
import math

def jump(array,target,value):
    step = math.sqrt(value)

    previous = 0
    while array[int(min(step, value)-1)] < target:
        previous = step
        step += math.sqrt(value)
        if previous >= value:
            return -1
        
    while array[int(previous)] < target:
        previous += 1

        if previous == min(step,value):
            return -1
    if array[int(previous)] == target:
        return previous
    return -1

array = [1,2,3,4,5,6,7,8,90,100]
target = 90
value = len(array)

index = jump(array, target, value)
print("number", target, "is at index", "%.0f"%index)
```
