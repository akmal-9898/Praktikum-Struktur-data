Simple Hash Table

```python
class hash_table:
  def __init__(self, size):
    self.size = size
    self.table = [[] for _ in range(self.size)]

  def function(self, key):
    return hash(key) % self.size

  def insert(self, key, value):
    index = self.function(key)
    self.table[index] = value

  def search(self, key):
    index = self.function(key)
    return self.table[index]

test = hash_table(6)
object1 = "amba"
object2 = "rusdi"
test.insert(object1, 100)
test.insert(object2, 200)

print(f"value for {object1} = {test.search(object1)}")
print(f"value for {object2} = {test.search(object2)}")
```

Handling collusion with Linear Probing

```python
class hash_table:
  def __init__(self, size):
    self.size = size
    self.table = [None] * size

  def function(self, key):
    return hash(key) % self.size

  def insert(self, key, value):
    index = self.function(key)
    origin_index = index

    while self.table[index] is not None:
      if self.table[index][0] == key:
        break
      index = (index + 1) % self.size
      if index == origin_index:
        raise Exception("Hash table is full")

    self.table[index] = (key, value)
  def search(self, key):
    index = self.function(key)
    origin_index = index

    while self.table[index] is not None:
      if self.table[index][0] == key:
        return self.table[index][1]
      index = (index + 1) % self.size
      if index == origin_index:
        break
    return None

test = hash_table(5)
object1 = "amba"
object2 = "rusdi"
object3 = "fuad"
test.insert(object1, 100)
test.insert(object2, 200)
test.insert(object3, 300)

print(f"value for {object1} = {test.search(object1)}")
print(f"value for {object2} = {test.search(object2)}")
print(f"value for {object3} = {test.search(object3)}")
```

Handling collusion using Chaining

```python
class hash_table:
  def __init__(self, size):
    self.size = size
    self.table = [[] for _ in range(size)]

  def function(self, key):
    return hash(key) % self.size

  def insert(self, key, value):
    index = self.function(key)
    for another_index, (another_key, another_value) in enumerate(self.table[index]):
      if another_key == key:
        self.table[index][another_index] = (key, value)
        return
    self.table[index].append((key, value))

  def search(self, key):
    index = self.function(key)
    for another_key, another_value in self.table[index]:
      if another_key == key:
        return another_value
    return None

test = hash_table(5)
object1 = "amba"
object2 = "rusdi"
object3 = "fuad"
test.insert(object1, 100)
test.insert(object2, 200)
test.insert(object3, 300)

print(f"value for {object1} = {test.search(object1)}")
print(f"value for {object2} = {test.search(object2)}")
print(f"value for {object3} = {test.search(object3)}")
```

1. Eksperiment sederhana Linear Probing

```python
class linear_probing:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.function(key)
        origin_index = index
        while self.table[index] is not None:
            if self.table[index][0] == key:
                break
            index = (index + 1) % self.size
            if index == origin_index:
                raise Exception("Hash table is full")
        self.table[index] = (key, value)

    def show(self):
        print("isi tabel :")
        for another_index, item in enumerate(self.table):
            print(f"Index {another_index} : {item}")

test = linear_probing(5)
test.insert("A", 10)
test.insert("B", 20)
test.insert("C", 30)
test.insert("D", 40)
test.insert("E", 50)
test.show()
```
