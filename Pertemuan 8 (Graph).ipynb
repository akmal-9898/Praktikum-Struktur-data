Adjacency Matrix

```python
class graph:
    def __init__(self, total_nodes):
        self.size = total_nodes     # Store the total number of nodes in an object variable
        self.adj_matrix = [[0] * total_nodes for _ in range(total_nodes)]       # Create a 2D (two-dimensional) matrix and fill it with zeros

    def edge(self, node1, node2):
        if node1 < self.size and node2 < self.size:     # Ensure that the number of vertices entered does not exceed the specified graph size
            self.adj_matrix[node1][node2] = 1       # Set the values in the matrix to 1
            self.adj_matrix[node2][node1] = 1       # Set the values in the matrix to 1

    def show(self):
        print(f"Adjacency Matrix ({self.size}x{self.size}):")
        for row in self.adj_matrix:     # Loop through each line in the
            print(row)

my_graph = graph(5)
my_graph.edge(0, 1)
my_graph.edge(0, 2)
my_graph.edge(1, 3)
my_graph.edge(2, 4)

my_graph.show()
```

Adjacency List

```python
class graph:
    def __init__(self):
        self.adjacency_list = {}        # stores a node (key) and a list of other nodes connected to it (value is a list).

    def edge(self, node1, node2):
        if node1 not in self.adjacency_list:
            self.adjacency_list[node1] = []     # Create a new container with the value of an empty list [].
        if node2 not in self.adjacency_list:
            self.adjacency_list[node2] = []     # Create a new container with the value of an empty list [].
        self.adjacency_list[node1].append(node2)        # add each other's nodes to their neighbor lists
        self.adjacency_list[node2].append(node1)        # add each other's nodes to their neighbor lists

    def show(self):
        for vertex in self.adjacency_list:      # Loop through each node (key)
            print(vertex, "->", " -> ".join(map(str, self.adjacency_list[vertex])))

my_graph = graph()
my_graph.edge(0, 1)
my_graph.edge(0, 2)
my_graph.edge(1, 3)
my_graph.edge(2, 4)

print("adjacency_list :")
my_graph.show()
```

Transpose Graph (using the adjacency list representation)

```python
def edge(adjacency, source, destination):
    adjacency[source].append(destination)       # Retrieve the list of neighbors for the "source" node, then append the "destination" node to it

def show(adjacency, vertex):
    for outer_index in range(vertex):       # Loop through the number of points present
        print(outer_index, "-> ", end="")
        for inner_index in range(len(adjacency[outer_index])):      # Iterate through that node to examine all the destination nodes (neighbors) connected to it.
            print(adjacency[outer_index][inner_index], end=" ")
        print()

def transposeGraph(adjacency, transpose, vertex):
    for outer_index in range(vertex):       # Iterate through the graph to check every source node in the “outer_index” of the original graph.
        for inner_index in range(len(adjacency[outer_index])):      # Iterate through each destination point “adjacency[outer_index][inner_index]” from that starting point.
            edge(transpose, adjacency[outer_index][inner_index], outer_index)       # the part of the code where the arrow direction is reversed

if __name__ == '__main__':
    vertex = 5
    adjacency = [[] for outer_index in range(vertex)]
    edge(adjacency, 0, 1)
    edge(adjacency, 0, 4)
    edge(adjacency, 0, 3)
    edge(adjacency, 2, 0)
    edge(adjacency, 3, 2)
    edge(adjacency, 4, 1)
    edge(adjacency, 4, 3)
    transpose = [[] for outer_index in range(vertex)]
    transposeGraph(adjacency, transpose, vertex)
    show(transpose, vertex)
```

Breadth First Search (BFS)

```python
from collections import deque
def bfs(adjlist, startnode, visited):
    queue = deque([startnode])      # Create a new queue named “queue” and immediately add “startnode” to it as the first element.
    visited.add(startnode)      # Mark “startnode” as a visited node by adding it to the “visited” set
    print(f"The search starts at {startnode} :", end=" ")
    while queue:
        currentNode = queue.popleft()       # Retrieve and remove the first node at the front of the queue, then store it in the variable “currentNode”
        print(currentNode, end=" ")
        neighbors = sorted(adjlist.get(currentNode, []))        # Retrieves all neighbors of “currentNode”. The function returns an empty list.
        for neighbor in neighbors:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

def main():
    adjlist = {}
    def edge(node1, node2):     # Create a local function inside the `main` function named “edge” to connect two nodes with an edge
        if node1 not in adjlist:
            adjlist[node1] = []     # If the “node1” node is not yet in the adjlist, create a new entry—an empty list—to hold its neighbors
        if node2 not in adjlist:
            adjlist[node2] = []     # If the “node2” node is not yet in the adjlist, create a new entry—an empty list—to hold its neighbors
        adjlist[node1].append(node2)
        adjlist[node2].append(node1)

    edge(0, 1)
    edge(0, 2)
    edge(1, 2)
    edge(1, 3)
    edge(2, 3)
    edge(2, 4)
    start_from = 0
    visited_set = set()
    bfs(adjlist, start_from, visited_set)
    print()
if __name__ == "__main__":
    main()
```

Depth First Search (DFS)

```python
from collections import defaultdict
class graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def edge(self, node1, node2):
        self.graph[node1].append(node2)

    def DFSUtil(self, node2, visited):
        visited.add(node2)      # Mark “node2” by adding it to the visited set so it won't be processed again later.
        print(node2, end=' ')
        for neighbour in self.graph[node2]:
            if neighbour not in visited:
                self.DFSUtil(neighbour, visited)        # The function will immediately call itself to enter and explore that neighbor first. 

    def DFS(self, node2):
        visited = set()     # Create an empty set named “visited” to keep track of the nodes that have been visited during the process.
        self.DFSUtil(node2, visited)

if __name__ == "__main__":
    my_graph = graph()
    my_graph.edge(0, 1)
    my_graph.edge(0, 2)
    my_graph.edge(1, 2)
    my_graph.edge(2, 0)
    my_graph.edge(2, 3)
    my_graph.edge(3, 3)
    the_vertex = 1
    print(f"DFS testing with vertex : {the_vertex}")
    my_graph.DFS(the_vertex)
```

Detect cycle in a directed graph

```python
from collections import defaultdict
class graph():
    def __init__(self, vertices):
        self.graph = defaultdict(list)      # Create a “self.graph” attribute using “defaultdict(list)”. This will store the graph representation as an adjacency list  
        self.V = vertices
        
    def edge(self, node1, node2):
        self.graph[node1].append(node2)     # Add “node2” to the list of neighbors for “node1”
        
    def isCyclicUtil(self, node2, visited, recstack):
        visited[node2] = True       # Mark the current node as True in the visited array because this node is currently being visited.
        recstack[node2] = True      # Set the current node to True in “recstack”
        for neighbour in self.graph[node2]:
            if visited[neighbour] == False:
                if self.isCyclicUtil(neighbour, visited, recstack) == True:     # The function will call itself recursively to check those neighbors
                    return True
            elif recstack[neighbour] == True:       # That neighbor is already in the active recursion stack
                return True
        recstack[node2] = False
        return False

    def isCyclic(self):
        visited = [False] * (self.V + 1)        # Create a “visited” list containing the value False for the number of nodes plus one to track which nodes have already been visited.
        recstack = [False] * (self.V + 1)       # Create a “recstack” list containing the value False as many times as the number of nodes plus one to track the recursion stack.
        for node in range(self.V):
            if visited[node] == False:
                if self.isCyclicUtil(node, visited, recstack) == True:
                    return True
        return False

if __name__ == '__main__':
    my_graph = graph(4)
    my_graph.edge(0, 1)
    my_graph.edge(0, 2)
    my_graph.edge(1, 2)
    my_graph.edge(2, 0)
    my_graph.edge(2, 3)
    my_graph.edge(3, 3)
    status = "graph"
    if my_graph.isCyclic() == 1:
        print(f"{status} = contains cycle")
    else:
        print(f"{status} = doesn't contain cycle")
```

kruskal algorithm (to find minimum spanning tree)

```python
class graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []

    def edge(self, origin_vertex, destination_vertex, burden):
        self.graph.append([origin_vertex, destination_vertex, burden])      # Function to add an edge between two vertices to “self.graph”

    def find(self, parent, index):
        if parent[index] == index:
            return index
        parent[index] = self.find(parent, parent[index])
        return parent[index]        # This technique directly connects the node to its group leader, making subsequent searches much faster.

    def union(self, parent, rank, own1, own2):      # combining groups with shorter trees with those with taller trees
        if rank[own1] < rank[own2]:
            parent[own1] = own2
        elif rank[own1] > rank[own2]:
            parent[own2] = own1
        else:
            parent[own2] = own1
            rank[own1] += 1     

    def KruskalMST(self):
        result = []
        index = 0
        edge_count = 0
        self.graph = sorted(self.graph, key=lambda item: item[2])       # sort all edges of the graph by their weights
        parent = []
        rank = []

        for node in range(self.V):      # Establish an initial condition in which each point is independent 
            parent.append(node)
            rank.append(0)
        while edge_count < self.V - 1:
            origin_vertex, destination_vertex, burden = self.graph[index]       # Select the side with the smallest weight at this point 
            index = index + 1
            own1 = self.find(parent, origin_vertex)     # find the group leader from the starting point
            own2 = self.find(parent, destination_vertex)        # find the group leader from the starting point
            if own1 != own2:        # Those two points are in different groups / do not form a cycle
                edge_count = edge_count + 1
                result.append([origin_vertex, destination_vertex, burden])
                self.union(parent, rank, own1, own2)

        minimumCost = 0
        print("Edges in the constructed MST")
        for origin_vertex, destination_vertex, weight in result:        # Iterate through the contents of “result” variable
            minimumCost += weight       # sum the total weights into the “minimumCost”
            print("%d -- %d == %d" % (origin_vertex, destination_vertex, weight))       # and then print the MST paths along with their total costs to the screen.
        print("Minimum :", minimumCost)

if __name__ == '__main__':
    my_graph = graph(4)
    my_graph.edge(0, 1, 10)
    my_graph.edge(0, 2, 6)
    my_graph.edge(0, 3, 5)
    my_graph.edge(1, 3, 15)
    my_graph.edge(2, 3, 4)
    my_graph.KruskalMST()
```

```python
def topologicalSortUtil(vertex, adjacency, visited, stack):
    visited[vertex] = True
    for neighbor in adjacency[vertex]:
        if not visited[neighbor]:
            topologicalSortUtil(neighbor, adjacency, visited, stack) 
    stack.append(vertex)

def topologicalSort(adjacency, total_vertices):
    stack = []
    visited = [False] * total_vertices
    for i in range(total_vertices):
        if not visited[i]:
            topologicalSortUtil(i, adjacency, visited, stack)
    print("sequence results :")
    print(stack[::-1])

if __name__ == "__main__":
    total_vertices = 6
    edges = [(5, 2), (5, 0), (4, 0), (4, 1), (2, 3), (3, 1)]
    adjacency = [[] for _ in range(total_vertices)]
    for source, destination in edges:
        adjacency[source].append(destination)
    topologicalSort(adjacency, total_vertices)
```
