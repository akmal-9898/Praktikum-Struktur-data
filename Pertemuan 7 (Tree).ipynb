```python
class node_at_binary_tree:
    def __init__(self,data):
        self.left = None        # Prepare left empty branches or sub-branches
        self.right = None       # Prepare right empty branches or sub-branches
        self.data = data     # Stores the value of the key entered in an internal variable
    def insert(self, data):
        if self.data:
            if data < self.data:        # If the new number is smaller than the number at the current node, it will try moving to the left.
                if self.left is None:       
                    self.left = node_at_binary_tree(data)       # The node will be placed there if the left node is empty
                else :
                    self.left.insert(data)      # If the node is already full, the insert function will be called until an empty slot is found
            elif data > self.data:      #  If the new number is smaller than the number at the current node, it will try moving to the right.
                if self.right is None:
                    self.right = node_at_binary_tree(data)      # # The node will be placed there if the right node is empty
                else :
                    self.right.insert(data)     # # If the node is already full, the insert function will be called until an empty slot is found
        else :
            self.data = data

    def show(self):
        if self.left:
            self.left.show()
        print(self.data, end=" "),
        if self.right:
            self.right.show()

way = node_at_binary_tree(10)
way.insert(20)
way.insert(90)
way.insert(50)
way.insert(40)
way.show()
```

In Order Transversal

```python
class node_at_binary_tree:
    def __init__(self, data):
        self.left = None         # Prepare left empty branches for smaller values
        self.right = None        # Prepare right empty branches for larger values
        self.data = data         # Stores the value of the key entered in an internal variable

    def insert(self, data):
        if self.data:       # Check if the current node already has data
            if data < self.data:        # If new data is smaller, move to the left subtree
                if self.left is None:
                    self.left = node_at_binary_tree(data)       # Create new node if left is empty
                else:
                    self.left.insert(data)      # Recursively call insert on left child
            elif data > self.data:      # If new data is larger, move to the right subtree
                if self.right is None:
                    self.right = node_at_binary_tree(data)      # Create new node if right is empty
                else:
                    self.right.insert(data)     # Recursively call insert on right child
        else:
            self.data = data     # Assign data to the current node if it was empty

    def show(self):
        if self.left:       # First, visit the left subtree
            self.left.show()
        print(self.data),       # Then, print the current node
        if self.right:      # Finally, visit the right subtree
            self.right.show()
    
    def order(self, way):
        res = []        # Initialize an empty list to store the results
        if way:     # If the current node (way) exists:
            res = self.order(way.left)
            res.append(way.data)
            res = res + self.order(way.right)
        return res 
    
way = node_at_binary_tree(90)
way.insert(50)
way.insert(70)
way.insert(80)
way.insert(60)
way.insert(40)
way.insert(30)
print(way.order(way))
```

Post Order Traversal

```python
class node_at_binary_tree:
    def __init__(self, data):
        self.left = None         # Prepare left empty branches for smaller values
        self.right = None        # Prepare right empty branches for larger values
        self.data = data         # Stores the value of the key entered in an internal variable

    def insert(self, data):
        if self.data:            # Check if the current node already has data
            if data < self.data:         # If new data is smaller, move to the left subtree
                if self.left is None:
                    self.left = node_at_binary_tree(data)       # Create new node if left is empty
                else:
                    self.left.insert(data)      # Recursively call insert on left child
            elif data > self.data:      # If new data is larger, move to the right subtree
                if self.right is None:
                    self.right = node_at_binary_tree(data)      # Create new node if right is empty
                else:
                    self.right.insert(data)     # Recursively call insert on right child
        else:
            self.data = data     # Assign data to the current node if it was empty

    def show(self):
        if self.left:            # visit the left subtree
            self.left.show()
        print(self.data),        # print the current node (In-order traversal style)
        if self.right:           
            self.right.show()

    def order(self, way):
        res = []        # Initialize an empty list to store the results
        if way:     # If the current node (way) exists:
            res = self.order(way.left)      # recursively visit all nodes in the left subtree
            res = res + self.order(way.right)       # recursively visit all nodes in the right subtree and combine results
            res.append(way.data)        # finally, add the current node's data (Left -> Right -> Root)
        return res           

way = node_at_binary_tree(50)     
way.insert(60)                    
way.insert(40)
way.insert(70)
way.insert(30)
way.insert(10)
way.insert(20)
way.insert(90)
way.insert(80)
print(way.order(way))
```

```python
class node_at_binary_tree:
    def __init__(self, data):
        self.left = None         # Prepare left empty branches or sub-branches
        self.right = None        # Prepare right empty branches or sub-branches
        self.data = data         # Stores the value of the key entered in an internal variable

def order(temp):
    if not temp:        # Base case: if node is empty, stop recursion
        return
    order(temp.left)        # Recursively visit the left branch
    print(temp.data, end=" ")       # Print current node data (In-order)
    order(temp.right)       # Recursively visit the right branch

def outermost_node(way, d_node):
    queue = []      # Create a queue for Level-Order traversal
    queue.append(way)       # Start with the root
    while(len(queue)):      # While there are nodes to check
        temp = queue.pop(0)     # Take the first node from the queue
        if temp is d_node:      # If the root itself is the node to delete
            temp = None
            return
        if temp.right:      # Check the right child first (deepest is usually right)
            if temp.right is d_node:
                temp.right = None       # Cut off the reference to the deepest node
                return
            else:
                queue.append(temp.right)
        if temp.left:            # Check the left child
            if temp.left is d_node:
                temp.left = None
                return
            else:
                queue.append(temp.left)

def delete(way, key):
    if way is None:     # If tree is empty, nothing to do
        return None
    if way.left is None and way.right is None:
        if way.data == key:
            return None 
        else:
            return way
    
    key_node = None     # Will store the node we want to delete
    queue = []
    queue.append(way)
    temp = None     # Will eventually store the deepest node
    
    while len(queue):
        temp = queue.pop(0)      # 'temp' will be the last node visited (deepest)
        if temp.data == key:
            key_node = temp      # Found the node to be deleted
        if temp.left:
            queue.append(temp.left)
        if temp.right:
            queue.append(temp.right)
            
    if key_node:        # If the target value was found in the tree
        x = temp.data       # Grab data from the deepest, rightmost node
        key_node.data = x       # Overwrite the target's data with 'x'
        outermost_node(way, temp)       # Delete the original deepest node
    return way

if __name__ == '__main__':
    root = node_at_binary_tree(10)
    root.left = node_at_binary_tree(40)
    root.left.left = node_at_binary_tree(70)
    root.left.right = node_at_binary_tree(20)
    root.right = node_at_binary_tree(90)
    root.right.left = node_at_binary_tree(50)
    root.right.right = node_at_binary_tree(80)
    
    print("before :", end=" ")
    order(root)
    key = 90
    root = delete(root, key)
    print("\nyour target :", key)
    print("after :", end=" ")
    order(root)
```

Search in Binary Tree

```python
class node_at_binary_tree:
    def __init__(self, key):
        self.key = key           # Store the value in the node
        self.left = None         # Initialize empty left child
        self.right = None        # Initialize empty right child

def insert(node, key):
    if node is None:          # If the current spot is empty, create and return a new node here
        return node_at_binary_tree(key)
    if key < node.key:          # If the key is smaller than the current node, go left
        node.left = insert(node.left, key)
    elif key > node.key:            # If the key is larger than the current node, go right
        node.right = insert(node.right, key)
    return node     # Return the (potentially updated) node pointer

def search(way, key):
    if way is None or way.key == key:           # if found the key or reached an empty leaf (not found)
        return way
    if way.key < key:       # If the key we want is larger than current, search the right side
        return search(way.right, key)
    return search(way.left, key)        # Osearch the left side

if __name__ == '__main__':
    way = None
    way = insert(way, 50)        
    insert(way, 30)              
    insert(way, 20)              
    insert(way, 40)              
    insert(way, 70)              
    insert(way, 60)              
    insert(way, 80)              
    key = 6
    if search(way, key) is None:
        print(key, "= not found")
    else:
        print(key, "= found")
    key = 60
    if search(way, key) is None:
        print(key, "= not found")
    else:
        print(key, "= found")
```

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def find_min_value_node(node):
    current = node
    while current.left is not None:
        current = current.left
    return current

def delete_node(root, key):
    if root is None:
        return root
    if key < root.key:
        root.left = delete_node(root.left, key)
    elif key > root.key:
        root.right = delete_node(root.right, key)
    else:
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        temp = find_min_value_node(root.right)
        root.key = temp.key
        root.right = delete_node(root.right, temp.key)
    return root

def inorder(root):
    if root:
        inorder(root.left)
        print(root.key, end=" ")
        inorder(root.right)

root = Node(50)
root.left = Node(30)
root.right = Node(70)
root.left.left = Node(20)
root.left.right = Node(40)
root.right.left = Node(60)
root.right.right = Node(80)
print(f"before :")
inorder(root)
print("\n\ndelete target : 20")
root = delete_node(root, 20)
inorder(root)
print("\n\ndelete target : 30")
root = delete_node(root, 30)
inorder(root)
print("\n\ndelete target : 50")
root = delete_node(root, 50)
inorder(root)
```
